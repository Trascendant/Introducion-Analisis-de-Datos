```python
import pandas as pd
import statistics

# Laboratorio #2 - 21 de agosto del 2023

# Integrantes:

# - Luis Fernández Serrano
# - Rafael Zúñiga Vindas
# - Daniel Gómez Mena
# - Emmanuel Pérez Sánchez

# Apertura de archivo
df=pd.read_csv('/content/casasboston.csv')
data=pd.DataFrame(df)

# 1. Seleccionar TOWN

# Seleccionando la ciudad de Lynn para realizar los puntos del 2 al 5
data_town_lynn = data.loc[data['TOWN'] == "Lynn"]

# 2. TOWN: CRIM aquellos casos que son superiores a 0.50

data_lynn_crim = data_town_lynn.loc[data_town_lynn['CRIM'] >= 0.50]
print("\nCrímen en Lynn superior a 0.50:\n")
print(data_lynn_crim)

# 3. Sacar el promedio de la edad, usar la función de promedio de matemática

promedio_data_edades = statistics.mean(data_town_lynn['AGE'])
print("\nPromedio de las edades en Lynn:\n")
print(promedio_data_edades)

# 4. Imprimir un índice y una edad, retorna los puntos correspondientes

indices_edad_stat_lynn = data_town_lynn.loc[(data_town_lynn['AGE'] >= 60) & (data_town_lynn['DIS'] > 4)]
print("\nAGE mayor o igual a 60 y DIS mayor que 4:\n")
print(indices_edad_stat_lynn)

# 5. Ordenamiento por CRIM (sort para el ordenamiento)

sort_data_lynn_crim = data_town_lynn.sort_values(by = ['CRIM'])
print("\nOrdenamiento ascendente por CRIM:\n")
print(sort_data_lynn_crim)
```
